```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('netflix_data.csv')

# Preprocessing
le = LabelEncoder()

for col in ['Gender', 'SubscriptionType', 'FavoriteGenre', 'SubscriptionRenewed']:
    df[col] = le.fit_transform(df[col])

# -----------------------------
# Q19: Target is MonthlySpend
# -----------------------------

# Features (X)
X_reg = df.drop(['UserID', 'MonthlySpend'], axis=1)

# Target (Y)
y_reg = df['MonthlySpend']

# Split data
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X_reg,
    y_reg,
    test_size=0.2,
    random_state=42
)

# Train model
lr = LinearRegression()

lr.fit(X_train_r, y_train_r)

# -----------------------------
# Q20: Predict for first user
# -----------------------------

new_user = X_test_r.iloc[[0]]

prediction = lr.predict(new_user)

print(f"Predicted Monthly Spend for new user: ₹{prediction[0]:.2f}")
```
